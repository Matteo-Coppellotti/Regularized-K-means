```python
# read memberships from application
memberships = pd.read_csv("application/output/runs/k_2/run_8/memberships.csv", header=None).transpose()

# import subjects ids fro the datasets
ids = pd.read_csv('application/data/z_map_SCHZ.csv', index_col=0)
ids = pd.concat([ids, pd.read_csv('application/data/z_map_CONTROL.csv', index_col=0)], axis=0, sort=False)
ids = ids.index.tolist()
```

```python
memberships
```

```python
ids
```

```python
memberships[memberships[0]==0].index.values
```

```python
# i need to extract the ids of the sujects corresponding to the indices above
ids_array = np.array(ids)
ids_list =ids_array[memberships[memberships[0]==0].index.values].tolist()
ids_list
```

```python
df['clusters'] = df.index.isin(ids_list).astype(int)
df
```

```python
means_clusters = df.select_dtypes(include=[np.number]).groupby("clusters").mean()
means_original = df.select_dtypes(include=[np.number]).mean()

# histogram of the means per column
import matplotlib.pyplot as plt
import seaborn as sns
for col in means_clusters.columns:
    plt.figure()
    sns.barplot(x=means_clusters.index, y=means_clusters[col])
    plt.axhline(y=means_original[col], color='r', linestyle='--')
    plt.title(col)
    plt.show()
```
